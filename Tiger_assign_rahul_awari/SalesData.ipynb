```python
#For all regions find number of salesman and total sales. Return as a dataframe with three columns - Region, salesmen_count and total_sales
dfnew=pd.DataFrame(df.groupby(['Region','SalesMan'])['Sale_amt'].sum())
dfnew
```

```python
listregion=df['Region'].unique()
listregion
```

```python
listsalesmancount=[]
listregionamount=[]
for i in range(0,len(listregion)):
    salesmancount=dfnew.loc[listregion[i]].count()
    regionamount=dfnew.loc[listregion[i]]['Sale_amt'].sum()
    listsalesmancount.append(salesmancount)
    listregionamount.append(regionamount)
listregionamount
```

```python
listsalesmancount
```

```python
list1=[]
for i in range(0,len(listregion)):
    temp=[]
    temp.append(listregion[i])
    temp.append(listsalesmancount[i])
    temp.append(listregionamount[i])
    list1.append(temp)
dffive=pd.DataFrame(list1,columns=['Region','SalesMan Count','Sales Amount'])
```

```python
#Create a dataframe with total sales as percentage for each manager. Dataframe to contain manager and percent_sales
listper=[]
totalsales=df['Sale_amt'].sum()
dfsix=pd.DataFrame(df.groupby('Manager')['Sale_amt'].sum())
listmanager=df['Manager'].unique()
for i in range(0,len(listmanager)):
    valper=dfsix.loc[listmanager[i]]['Sale_amt']
    ans=(valper/totalsales)*100
    listper.append(ans)
dffin=pd.DataFrame(list(zip(listmanager,listper)),columns =['Manager', 'Percentage'])
dffin
```
