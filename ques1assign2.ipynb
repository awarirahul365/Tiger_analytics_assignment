```python
def func(df):
    df['flavor']=df['flavor'].astype(str)
    df.dropna(subset=['flavor'])
    len(df['flavor'])
    listflavor=[]
    for i in range(0,len(df['flavor'])):
        listflavor.append(df['flavor'][i])
    listtemp=[]
    for i in range(0,len(listflavor)):
        strtemp=listflavor[i].replace("||",";")
        listtemp.append(strtemp)
    listtemp.remove("nan")
    setflavor=set()
    for i in range(0,len(listtemp)):
        if listtemp[i]!="nan":
            str_ext=listtemp[i].split(";")
            for j in range(0,len(str_ext)):
                str_ele=str_ext[j].split(',')
                if len(str_ele)==1:
                    setflavor.add(str_ele[0].lower())
                else:
                    str1=str_ele[0].lower()
                    str2=str_ele[1].lower()
                    if(str2 == " not specified"):
                        setflavor.add(str1)
                    else:
                        setflavor.add(str2+str1)          
    return len(setflavor)
```
