```python
df_fruit=pd.read_csv(r"C:\Users\Rahul Awari\Downloads\Assignment\Flavor Classification Dataset.csv",encoding='latin-1')
df_fruit
```

```python
dftemp=df_fruit[(df_fruit['Flavor_Group'] == "Fruit")]
dftemp['flavor'].astype(str)
dftemp
```

```python
listfruit=dftemp['flavor'].tolist()
listfruit
```

```python
df['eventdate']=pd.to_datetime(df['eventdate'])
df
```

```python
df['quarter']=df['eventdate'].dt.to_period('Q')
```

```python
listquarter=df['quarter'].unique()
len(listquarter)
```

```python
list1=[]
for i in listquarter:
    count=0
    for j in listfruit:
        v=len(df[(df['quarter'] == i) & (df['flavor'].str.match(j,case=False))])
        count=count+v
    list1.append(count)
list1
```
