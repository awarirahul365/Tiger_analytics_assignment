```python
df['eventdate']=pd.to_datetime(df['eventdate'])
df['year']=df['eventdate'].dt.year
df
```

```python
listyear=df['year'].unique().tolist()
listyear
```

```python
listethical=df_ethical['Positioning Subcategory'].tolist()
listethical
```

```python
df[(df['year'] == 2006) & (df['positioning'].str.contains("Ethical",case=False))]
```

```python
listeth=[]
listcon=[]
listethcon=[]
sdup=set([])
for i in range(0,len(listyear)):
    eth1=len(df[(df['year'] == listyear[i]) & (df['positioning'].str.contains("Ethical",case=False)) & (~df['positioning'].str.contains("Convenience", na=False))])
    con1=len(df[(df['year'] == listyear[i]) & (df['positioning'].str.contains("Convenience",case=False)) & (~df['positioning'].str.contains("Ethical", na=False))])
    ethcon=len(df[(df['year'] == listyear[i]) & (df['positioning'].str.contains("Convenience",case=False)) & (df['positioning'].str.contains("Ethical",case=False))])
    listeth.append(eth1)
    listcon.append(con1)
    listethcon.append(ethcon)
listeth
```

```python
listcombine=[]
listcombine.append(listeth)
listcombine.append(listcon)
listcombine.append(listethcon)
listcombine
```

```python
df_individual=pd.DataFrame(listcombine,index=['Only Ethical','Only Convenience','Convenience & Ethical Overlap'],columns=[2005, 2009, 2006, 2008, 2007, 2010, 2011, 2012, 2013, 2014])
df_individual
```
