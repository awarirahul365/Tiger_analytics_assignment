```python
df['eventdate']=pd.to_datetime(df['eventdate'])
df
```

```python
df['quarter']=df['eventdate'].dt.to_period('Q')
df.head()
```

```python
df['year']=df['eventdate'].dt.year
df
```

```python
df['monthly']=df['year'].astype(str)+"M"+df['eventdate'].dt.month.astype(str)
df
```

```python
listmarket=df['market_subcategory'].unique().tolist()
listmarket
```

```python
listquarter=df['quarter'].unique()
len(listquarter)
```

```python
listans=[[]]
for i in range(0,len(listquarter)):
    listtemp=[]
    for j in range(0,len(listmarket)):
        v=len(df[(df['quarter'] == listquarter[i]) & (df['market_subcategory'] == listmarket[j])])
        listtemp.append(v)
    listans.append(listtemp)
listans.pop(0)
```
