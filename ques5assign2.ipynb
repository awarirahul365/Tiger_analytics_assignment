```python
df3=df[(df['date'].dt.year == 2013)]
df3
```

```python
listcountry=df['country'].unique()
listcountry
```

```python
listtemp=[[]]
for cont in listcountry:
    df8=df3[(df3['country'] == cont)]
    df8['flavor']=df8['flavor'].astype(str)
    df8.dropna(subset=['flavor'])
    listflv=df8['flavor'].tolist()
    list2=[]
    for i in range(0,len(listflv)):
        strtemp=listflv[i].replace("||",";")
        list2.append(strtemp)
    freq={}
    for i in range(0,len(list2)):
        if list2[i]!="nan":
            stl=list2[i].split(';')
            for j in range(0,len(stl)):
                slw=stl[j].lower()
                z=slw.replace(" ","")
                if z in freq:
                    freq[z]=freq[z]+1
                else:
                    freq[z]=1
    dict_sorted=sorted(freq,key=freq.get,reverse=True)
    listfive=[]
    for i in dict_sorted:
        listfive.append(i)
    if len(listfive)<5:
        listtemp.append(listfive)
    else:
        listtemp.append(listfive[:5])
listtemp.pop(0)
len(listtemp)
```

```python
hm=dict(zip(listcountry,listtemp))
hm
```

```python
pd.DataFrame.from_dict(hm,orient='index')
```
