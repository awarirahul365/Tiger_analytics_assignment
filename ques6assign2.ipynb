```python
list1=df_cat['Positioning Group'].unique()
list3=df_cat['Positioning Subcategory'].unique()
list3
```

```python
list2=[]
for i in range(0,len(list1)):
    listtemp=[]
    for j in range(0,len(list3)):
        v=len(df_cat[(df_cat['Positioning Group'] == list1[i]) & (df_cat['Positioning Subcategory'] == list3[j])])
        if v>0:
            listtemp.append(list3[j])
    list2.append(listtemp)
list2
```

```python
listans=[]
for i in range(0,len(list2)):
    count=0
    for j in range(0,len(list2[i])):
        z=len(df_2013[(df_2013['positioning'] == list2[i][j] )])
        count=count+z
    listans.append(count)
listans
```
